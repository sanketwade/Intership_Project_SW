```python
from pyspark.sql import SparkSession

# Spark session initialize 
spark = SparkSession.builder.appName("HR_Staging_Layer").getOrCreate()

# file paths
source_file_path = injection_lh_fpath+"//"+"HR_DATA.csv"  # Source CSV file
staging_file_path = injection_lh_fpath+"//"+"hr_data_staging.csv"  # Staging layer file

# Load the raw HR data into a DataFrame
hr_staging_df = spark.read.option("header", "true").option("inferSchema", "true").csv(source_file_path)

# Save the raw data into the Staging Layer as CSV

hr_staging_df.write.mode("overwrite").option("header", "true").csv(staging_file_path)
# hr_staging_df.show(5)

# display(hr_staging_df)

print("HR data successfully loaded into the Staging Layer: 'hr_data_staging.csv'")
```
